### Temperature related variables df

```python
temp_var = ['temperature','apparentTemperature', 'short_summary', 'precipIntensity','precipProbability', 'humidity', 
        'windSpeed', 'visibility','temperatureHigh', 'temperatureLow', 'dewPoint', 'windBearing','cloudCover', 
          'uvIndex', 'visibility.1', 'moonPhase', 'price']
```

```python
temp_df = both[temp_var]
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(temp_df.corr()[['price']].sort_values('price'),
 vmax=1, vmin=-1, cmap='coolwarm', annot=True, ax=ax);
ax.invert_yaxis()
```

```python
sns.scatterplot(x= 'temperature', y = 'apparentTemperature', data = temp_df);
```
