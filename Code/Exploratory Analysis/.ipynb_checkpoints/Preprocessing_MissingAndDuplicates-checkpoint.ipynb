## Missing Data

Missing data? Yes. 1 track name, 1 artist and 1,472 genres. Will get rid of the row with 1 missing track and 1 missing artist but have to think about the genres as it's around 9% of my data.

```python
df.isna().sum()
```

Let's find where the track name is missing

```python
#removing the one track_name NA
df = df[pd.notnull(df['track_name'])]
#removing the one artist NA
df = df[pd.notnull(df['artist'])]
```

```python
# the same row had the missing track name and missing artist so we only lost 1 row
df.shape
```

Looking at which category has the most missing genres--remember 'genre' is more specific where 'category' is broad

```python
# create a dataframe with only category and genre
df_na = df[['category', 'genre']]
```

```python
df_na.head()
```

```python
# creating a boolean column with 1's and 0's depending if the genre is null or not
df_na['genre_nulls'] = df_na['genre'].isnull().astype(int)
```

```python
# create a df from the above dataframe that aggregates the null values by category
df_agg = df_na.groupby(['category']).agg(['sum','count'])
```

```python
# adding column for percentage of nulls within each category
df_agg[('genre_nulls','na_percent')] = (df_agg['genre_nulls']['sum']) / (df_agg['genre_nulls']['count'])
```

```python
df_agg
```

Seems that the 'club' category had the most null values, followed by 'rap'. We want to plot percentage of missing values within each category

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(8, 8))

# Plot the total crashes
sns.set_color_codes("pastel");
missing_nulls = sns.barplot(x=df_agg.index, y=df_agg['genre_nulls']['count'], data=df_agg,
            label="Total", color="b");

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted");
sns.barplot(x=df_agg.index, y=df_agg['genre_nulls']["sum"], data=df_agg,
            label="Null Values", color="b");

# Add a legend and informative axis label
ax.legend(ncol=2, loc="upper right", frameon=True);
ax.set(ylabel="Number of Observations",
       xlabel="Category");
ax.set_title('Percentage of Null Values for each Category');

fig1 = missing_nulls.get_figure()
fig1.set_size_inches(11.7, 8.27)
fig1.savefig("../../Images_Plots/missing_nulls.png")
```

## Duplicates

Motivation: Being popular in one category is great but if your song is used in multiple genres and many playlists...eventually means it's more lucrative. More appearances in playlists --> more listens --> more popular --> more money

### Looking at duplicate songs

To deal with duplicates make a duplicates dataframe and aggregate similar to above. We want to look at the crossovers of songs in playlists.

```python
# Create df for looking at duplicates
df_duplicates = df[['track_ids', 'track_name', 'playlist_id', 'playlist_name', 'category']]
```

```python
# We can see there are definitely duplicates bc track_ids are appearing more than once
# This can be bc a myriad of reasons-- used in different playlists mainly, somehow we want to narrow down on this crossover
df_duplicates['track_ids'].value_counts()
```

```python
# Make a column that has how many times the song appeared
df_duplicates['duplicate_counts_ids'] = df_duplicates.groupby(['track_ids'])['category'].transform('count')
```

```python
df_duplicates.head()
```

```python
#checking to see if the column is correct.
# this code should produce 2 rows
df_duplicates[df_duplicates['track_ids'] == '3GJ4hzg4lrGwU51Y3VARbF']
```

```python
#This is suspicious bc 'Speechless is a pretty popular song'...check similar way but with track name
df_duplicates[df_duplicates['track_name'] == 'Speechless']
```

For this particular track, the 'track_name' encaptures the correct subset of data

```python
df_duplicates['duplicate_counts_names'] = df_duplicates.groupby(['track_name'])['category'].transform('count')
```

```python
df_duplicates.head()
```

```python
# This plot shows that the names count is always greater than or equal to the ids count
# Note that there are way fewer than 16,000 something dots but that's cuz there are a lot of the same combos within data
plt.figure(figsize= (10,7))
track_duplicates = sns.scatterplot(x= 'duplicate_counts_ids', y = 'duplicate_counts_names', data = df_duplicates);
plt.title('Track IDs vs Track Names')
fig2 = track_duplicates.get_figure()
fig2.set_size_inches(11.7, 8.27)
fig2.savefig("../../Images_Plots/track_duplicates.png")
```

```python
def three_matches(set1, set2, set3):
    temp=[]
    for i in set1:
        if (i in set2) and (i in set3):
            temp.append(1)
            print(i)

    print(sum(temp))
```

```python
def four_matches(set1, set2, set3, set4):
    temp=[]
    for i in set1:
        if (i in set2) and (i in set3) and (i in set4):
            temp.append(1)

    print(sum(temp))
```
