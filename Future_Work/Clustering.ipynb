```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_csv('../Data/final_without_duplicates.csv')
df.head()
```

## We only want to use the audio features (maybe categorical variables later)

Did not use duration or time signature

```python
df.columns
```

```python
df = df[['track_name', 'artist', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'time_signature', 'category']]
```

```python
X = df.loc[:, ['danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo']]
```

Have to scale the data because tempo and other features are not on the same scale (0-1)

```python
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
km = KMeans(n_clusters=100, random_state=42)
km.fit(X_sc)
```

```python
df['cluster'] = km.predict(X_sc)
df.head()
```

```python
df[df['cluster']==6]
```

```python
#check silhouette score-want higher score
silhouette_score(X_sc, km.labels_)
```

## What if I used the genre?

```python
temp = df['category']
print(temp.head())
temp = pd.get_dummies(temp)
temp
```

```python
df.drop('category', axis =1, inplace =True)
df.columns
```

```python
df = pd.concat([df, temp], axis=1, sort=False)
```

```python
df.columns
```

```python
X2 = df.loc[:, ['danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'alternative', 'club', 'country', 'hits', 'pop', 'rap']]
```

```python
X2
```

```python
sc2 = StandardScaler()
X2_sc = sc2.fit_transform(X2)
```

```python
km2 = KMeans(n_clusters=100, random_state=42)
km2.fit(X2_sc)
```

```python
df['cluster2'] = km2.predict(X2_sc)
df.head()
```

The score improves very slightly as seen below when we add the genre

```python
silhouette_score(X2_sc, km2.labels_)
```
